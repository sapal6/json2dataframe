```python
#todo: the rmtree line tries to remove the kaggle working dir instead of that try to remove the file if its existsing
def untar(source:Path, dest:Path=None, data=True, force_download=False, verbose=False) -> Path:
    "Download `url` to `fname` if `dest` doesn't exist, and un-tgz to folder `dest`."
    if verbose: print(f'source:{source} dest:{dest}')
    dest = source if dest is None else dest/source.name #todo: set the target file to dest/soruce's fiel name
    #if dest.exists(): shutil.rmtree(dest.name)
    if not dest.exists():
        if verbose: print("extracting")
        tarfile.open(source, 'r:gz').extractall(dest)
        if verbose: print('File extracted successfully.')
    return dest
```
